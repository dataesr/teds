```python
import os
import requests
import pandas as pd
```

```python
import sys
sys.path.append('C:\\Users\\haallat\\Documents\\teds\\code_utils')  

from elastic import get_from_es
```

```python
body = {
        "size": 0,
        "query": {
            "bool": {
                "must": [{"term": {"tags.id.keyword": "ipcc_wg1"}}],
            },
        },
        "aggs": {
            "co_authors": {
                "terms": {"field": f"co_authors.keyword", "size": 20000000},
            },
        },
    }
get_from_es(body,'scanr-publications')
```

```python
res = get_from_es(body,'scanr-publications')
co_authors = res.get("aggregations").get("co_authors").get("buckets")
```

```python
links = []
for link in co_authors:
    source = link["key"].split("---")[0].split("###")[1]
    target = link["key"].split("---")[1].split("###")[1]
    links.append({"source_id": source, "target_id": target, "strength": link["doc_count"]})


links
```

```python
df = pd.DataFrame(links)
df.to_json("teds_ipcc_wg1_authors.json", orient="records")
```
